```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(1, 20, 100).reshape(-1, 1)
y = 10 * np.log(x) + np.random.normal(0, 0.5, size=x.shape[0])
model=RandomForestRegressor(n_estimators=10)
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.25)
model.fit(x_train,y_train)
pred=model.predict(x_test)
print(r2_score(y_test,pred))
```

```python
x = np.linspace(1, 20, 100)  # shape: (100,)

# Step 2: Generate y as a 1D array of the same length
y = 10 * np.log(x) + np.random.normal(0, 0.5, size=x.shape[0])  # shape: (100,)

# Step 3: Plot
plt.figure(figsize=(8, 5))
plt.scatter(x, y, color="blue", label="Synthetic data", alpha=0.7)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Synthetic dataset: y = 10 * log(x) + noise")
plt.legend()
plt.grid(True)
plt.show()
```
