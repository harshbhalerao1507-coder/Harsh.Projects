```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import preprocessing
from sklearn.metrics import r2_score,mean_squared_error
import matplotlib.pyplot as plt
```

```python
x=np.array([[1,2],[5,4],[7,8],[8,9],[9,7],[5,4],[7,8],[8,9],[9,7]])
y=np.array([1,5,6,5,6,4,8,7,9])
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.25)
model=LinearRegression()
model.fit(x_train,y_train)
pred=model.predict(x_test)
pred

y_pred = model.predict(x)
print("MSE:", mean_squared_error(y, y_pred))
print("R² Score:", r2_score(y, y_pred))




```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# 3D scatter plot of training data
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection="3d")
ax.scatter(x[:,0], x[:,1], y, color="blue", label="Actual y")

# Create a mesh grid for feature space
x1_range = np.linspace(x[:,0].min(), x[:,0].max(), 20)
x2_range = np.linspace(x[:,1].min(), x[:,1].max(), 20)
x1_mesh, x2_mesh = np.meshgrid(x1_range, x2_range)

# Predict values on the grid
y_mesh = model.predict(np.c_[x1_mesh.ravel(), x2_mesh.ravel()])
y_mesh = y_mesh.reshape(x1_mesh.shape)

# Plot regression plane
ax.plot_surface(x1_mesh, x2_mesh, y_mesh, color="red", alpha=0.5)

ax.set_xlabel("Feature 1")
ax.set_ylabel("Feature 2")
ax.set_zlabel("Target y")
plt.title("Multiple Linear Regression Plane")
plt.legend()
plt.show()
```
