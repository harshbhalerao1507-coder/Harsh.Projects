```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import preprocessing

```

```python
df=pd.read_csv("Lung cancer.csv")
df.head()
```

```python
df_binary=df[['bmi','cholesterol_level']].copy()
df_binary.head()
```

```python
sns.lmplot(x='bmi',y='cholesterol_level',data=df_binary,order=2,ci=None)
```

```python
x=np.array(df_binary['bmi']).reshape(-1,1)
y=np.array(df_binary['cholesterol_level']).reshape(-1,1)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25)
```

```python
reg=LinearRegression()
reg.fit(x_train,y_train)
```

```python
pred=reg.predict(x_test)
pred
```

```python
plt.scatter(x_test.flatten(),y_test.flatten())
plt.plot(x_test,pred,color='k')
plt.show()
```
